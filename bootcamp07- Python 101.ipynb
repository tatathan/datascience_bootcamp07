```python
#final exam
def grade(score):
    if score >=120:
        return "Excellent"
    elif score >= 100:
        return "Good"
    elif score >= 80:
        return "Okay"
    else:
        return "need to rea more"
```

```python
result = grade(1205)
print(result)
```

```python
#use and, or in condition
def grad(course, score):
    if course == "english" and score >= 70:
        return "passed"
    elif course == "data science" and score >= 80:
        return "passed"
    else:
        return "failed"
```

```python
grad("data science", 100)
```

```python
#for loop

def grading_all(scores):
    new_scores = []
    for score in scores:
        new_scores.append(score+2)
    return new_scores
```

```python
grading_all([75, 88 ,90, 95, 53])
```

```python
grading_all((75, 88 ,90, 95, 53))
```

```python
grading_all({75, 88 ,90, 95, 53})
```

```python
#dog class
class Dog:
    def __init__(self, name, age, breed):
        self.name = name
        self.age = age
        self.breed = breed
```

```python
dog1= Dog("ovaltine", 2, "chihuahua")
dog2= Dog("milo", 3, "bulldog")
dog3= Dog("pepsi", 3.5, "german shepherd")
```

```python
print(dog1.name, dog1.age, dog1.breed)
print(dog2.name, dog2.age, dog2.breed)
print(dog3.name, dog3.age, dog3.breed)
```

```python
dog4= Dog("wick", 4, "assasin") #ลอคไว้แล้วว่า Dog(name,age,breed)
```

```python
#กำหนด method ได้ด้วย
class Employee:
    def __init__(self, id, name, dept, pos):
        self.id = id
        self.name = name
        self.dept = dept
        self.pos = pos
    def hello(self):
        print( f"Hello! my name is {self.name}")
    def work_hour(self, hours):
        print( f"{self.name} work for {hours} hours.") #hours ไม่ใช่ค่าใน
    def change_dept(self, new_dept):
        self.dept = new_dept
        print( f"new dept is {self.dept}")
```

```python
emp1 = Employee(1, "john", "Finance", "Financial Analyst")
print(emp1.name, emp1.pos)
```

```python
emp1.hello()
```

```python
emp1.work_hour(10)
```

```python
emp1.change_dept("Data Science")
```

```python
#object: attribute => name, id, dept
        #method => hello()
```
